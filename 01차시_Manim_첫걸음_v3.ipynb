## 1. 첫 애니메이션 실행해 보기

Manim 코드는 항상 이런 구조예요.

```
%%manim -qm -v WARNING 장면이름     ← Colab에게 "애니메이션 만들어줘"라고 알리는 명령
class 장면이름(Scene):              ← 하나의 애니메이션(장면)
    def construct(self):           ← 장면 안에서 일어날 일을 적는 곳
        ...                        ← 도형 만들기 + 움직이기
```

아래 셀을 실행해서 점과 원이 그려지는 애니메이션을 만들어 보세요.

````python
%%manim -qm -v WARNING DrawCircleWithPoint

class DrawCircleWithPoint(Scene):
    def construct(self):
        dot = Dot(point=ORIGIN, color=ORANGE)      # 원점에 주황색 점
        circle = Circle(radius=2, color=WHITE)     # 반지름 2인 하얀 원

        self.play(FadeIn(dot))       # 점이 서서히 나타남
        self.wait(0.5)
        self.play(Create(circle))    # 원이 그려짐
        self.wait(1)
````

## 2. 여러 가지 도형 만들기

Manim에는 수학 도형이 미리 준비되어 있어요.

| 코드 | 도형 |
|---|---|
| `Dot()` | 점 |
| `Line(시작점, 끝점)` | 선분 |
| `Circle(radius=2)` | 원 |
| `Square(side_length=2)` | 정사각형 |
| `Triangle()` | 정삼각형 |
| `Polygon(점1, 점2, 점3, ...)` | 다각형 |

위치는 `shift()`로 옮겨요: `UP`(위), `DOWN`(아래), `LEFT`(왼쪽), `RIGHT`(오른쪽)

````python
%%manim -qm -v WARNING ManyShapes

class ManyShapes(Scene):
    def construct(self):
        circle = Circle(radius=1, color=BLUE).shift(LEFT * 4)
        square = Square(side_length=2, color=GREEN)
        triangle = Triangle(color=RED).shift(RIGHT * 4)

        self.play(Create(circle))
        self.play(Create(square))
        self.play(Create(triangle))
        self.wait(1)
````

## 3. 색 바꾸기, 채우기, 글자 넣기

- 색: `RED, ORANGE, YELLOW, GREEN, BLUE, PURPLE, PINK, WHITE` ...
- 채우기: `set_fill(색, opacity=0.5)` — opacity는 0(투명)~1(불투명)

> 💡 **한글 글자 쓰기**: Manim에서 한글을 쓰려면 `Text("내용", font="NanumGothic")` 처럼 글꼴을 지정해야 해요.
> 수업 내내 자주 쓰니까 기억해 두세요!

````python
%%manim -qm -v WARNING ColorAndText

class ColorAndText(Scene):
    def construct(self):
        title = Text("나의 첫 Manim 작품", font="NanumGothic", font_size=40)
        title.to_edge(UP)                         # 화면 위쪽 가장자리로

        hexagon = Polygon(                        # 점 6개로 육각형 만들기
            [2, 0, 0], [1, 1.7, 0], [-1, 1.7, 0],
            [-2, 0, 0], [-1, -1.7, 0], [1, -1.7, 0],
            color=PURPLE
        )
        hexagon.set_fill(PURPLE, opacity=0.5)     # 보라색으로 반투명하게 채우기

        self.play(Write(title))                   # 글자가 써지는 애니메이션
        self.play(Create(hexagon))
        self.play(hexagon.animate.rotate(PI / 6)) # 30도 회전 (PI = 180도)
        self.wait(1)
````
