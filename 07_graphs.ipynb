## 1. 정비례 y = ax 그래프

`Axes`(좌표축)를 만들고 `plot`으로 그래프를 그려요.
y = 2x : x가 1 늘면 y는 2씩 늘어나는 **원점을 지나는 직선**!

```python
%%manim -qm -v WARNING DirectProportion

class DirectProportion(Scene):
    def construct(self):
        axes = Axes(x_range=[-4, 4, 1], y_range=[-4, 4, 1],
                    x_length=7, y_length=7)
        self.play(Create(axes))

        # y = 2x 그래프
        graph1 = axes.plot(lambda x: 2 * x, x_range=[-2, 2], color=YELLOW)
        t1 = Text("y = 2x", font_size=32, color=YELLOW)
        t1.next_to(axes.c2p(1.7, 3.4), RIGHT)
        self.play(Create(graph1), Write(t1))

        # 표처럼 점도 찍어 보기: x = -2, -1, 0, 1, 2
        for x in [-2, -1, 0, 1, 2]:
            d = Dot(axes.c2p(x, 2 * x), color=ORANGE)
            self.play(FadeIn(d), run_time=0.3)

        # a가 음수면? y = -x
        graph2 = axes.plot(lambda x: -x, x_range=[-3.5, 3.5], color=GREEN)
        t2 = Text("y = -x", font_size=32, color=GREEN)
        t2.next_to(axes.c2p(-3, 3), LEFT)
        self.play(Create(graph2), Write(t2))
        self.wait(2)
```

## 2. a가 변하면 그래프는? (움직이는 그래프!)

`ValueTracker`는 **변하는 수**를 담는 통이에요.
a가 0.5 → 3 → -2 로 변할 때 직선이 어떻게 기울어지는지 관찰하세요.

```python
%%manim -qm -v WARNING ChangingSlope

class ChangingSlope(Scene):
    def construct(self):
        axes = Axes(x_range=[-4, 4, 1], y_range=[-4, 4, 1],
                    x_length=7, y_length=7)
        self.play(Create(axes))

        a = ValueTracker(0.5)            # a의 시작값

        # a가 변할 때마다 그래프를 다시 그림
        graph = always_redraw(lambda: axes.plot(
            lambda x: a.get_value() * x, x_range=[-3.5, 3.5], color=YELLOW))

        label = always_redraw(lambda: Text(
            f"a = {a.get_value():.1f}", font_size=36, color=YELLOW
        ).to_corner(UR))

        self.add(graph, label)        # 자동 갱신되는 도형은 add로 올려요
        self.wait(0.5)
        self.play(a.animate.set_value(3), run_time=3)    # a: 0.5 → 3
        self.wait(0.5)
        self.play(a.animate.set_value(-2), run_time=3)   # a: 3 → -2
        self.wait(2)
```

## 3. 반비례 y = a/x 그래프

y = 6/x : x가 2배가 되면 y는 ½배!
그래프는 **두 개의 매끄러운 곡선(쌍곡선)** 이고, x = 0에서는 그릴 수 없어요. (0으로 나눌 수 없으니까!)

```python
%%manim -qm -v WARNING InverseProportion

class InverseProportion(Scene):
    def construct(self):
        axes = Axes(x_range=[-6, 6, 1], y_range=[-6, 6, 1],
                    x_length=7, y_length=7)
        self.play(Create(axes))

        # x=0 근처는 피해서 두 곡선으로 나눠 그려요
        right = axes.plot(lambda x: 6 / x, x_range=[1.1, 5.5], color=YELLOW)
        left = axes.plot(lambda x: 6 / x, x_range=[-5.5, -1.1], color=YELLOW)
        t = Text("y = 6/x", font_size=32, color=YELLOW)
        t.next_to(axes.c2p(2.5, 3), RIGHT)

        self.play(Create(right), Write(t))
        self.play(Create(left))

        # (1,6), (2,3), (3,2), (6,1) — 곱이 항상 6!
        for x, y in [(1.2, 5), (2, 3), (3, 2), (5, 1.2)]:
            self.play(FadeIn(Dot(axes.c2p(x, y), color=ORANGE)), run_time=0.3)

        msg = Text("x × y = 항상 6 (일정)", font="NanumGothic",
                   font_size=30, color=ORANGE).to_edge(DOWN)
        self.play(Write(msg))
        self.wait(2)
```
