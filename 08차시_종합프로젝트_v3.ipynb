### 주제 예시 (이 중에서 골라도, 새로 정해도 OK)

| 단원 | 아이디어 |
|---|---|
| 소인수분해 | 좋아하는 수(생일 등)의 소인수분해 트리 |
| 정수와 유리수 | 수직선 위 덧셈·뺄셈 문제 영상 (친구가 풀게!) |
| 기본 도형 | 시계 각도 퀴즈, 맞꼭지각 설명 영상 |
| 작도 | 정삼각형/정육각형 작도 과정 영상 |
| 좌표평면 | 좌표로 그린 나만의 캐릭터 |
| 그래프 | a가 변하는 움직이는 그래프 쇼 |

### 작품 조건 (체크리스트)

- [ ] 애니메이션이 3개 이상 있다 (`Create`, `FadeIn`, `animate` 등)
- [ ] 마무리 장면이 있다

## 2. 그동안 배운 도구 모음 (치트시트)

```python
# ── 도형 ──────────────────────────────
Dot([x, y, 0], color=RED)              # 점
Line(점1, 점2)                          # 선분
Arrow(점1, 점2, buff=0)                 # 화살표
Circle(radius=2) / Square() / Triangle() / Polygon(p1, p2, p3)
Angle(선1, 선2, radius=0.7)             # 각 표시
Text("한글", font="NanumGothic")        # 글자

# ── 수학 도구 ──────────────────────────
NumberLine(x_range=[-6, 6, 1], length=12)   # 수직선, line.n2p(수)
NumberPlane(x_range=..., y_range=...)        # 좌표평면, plane.c2p(x, y)
Axes(...) + axes.plot(lambda x: 2*x)         # 그래프

# ── 애니메이션 ─────────────────────────
self.play(Create(도형))         # 그려짐
self.play(FadeIn(도형))         # 나타남
self.play(Write(글자))          # 써짐
self.play(도형.animate.shift(RIGHT * 2))     # 이동
self.play(도형.animate.set_color(YELLOW))    # 색 변경
self.play(도형.animate.rotate(PI / 4))       # 회전
self.wait(1)                    # 기다리기

# ── 움직이는 값 ────────────────────────
a = ValueTracker(1)
graph = always_redraw(lambda: axes.plot(lambda x: a.get_value() * x))
self.play(a.animate.set_value(3), run_time=3)
```

## 3. 작품 템플릿

아래 템플릿의 3막 구조(제목 → 내용 → 마무리)를 채워 보세요.

````python
%%manim -qm -v WARNING MyProject

class MyProject(Scene):
    def construct(self):
        # ===== 1막: 제목 =====
        title = Text("작품 제목을 쓰세요", font="NanumGothic", font_size=48)
        author = Text("만든 사람: 홍길동", font="NanumGothic", font_size=28)
        author.next_to(title, DOWN)
        self.play(Write(title), FadeIn(author))
        self.wait(1)
        self.play(FadeOut(title), FadeOut(author))

        # ===== 2막: 내용 =====
        # 여기에 수학 개념 애니메이션을 만드세요!



        # ===== 3막: 마무리 =====
        ending = Text("감사합니다!", font="NanumGothic", font_size=48, color=YELLOW)
        self.play(Write(ending))
        self.wait(2)
````
