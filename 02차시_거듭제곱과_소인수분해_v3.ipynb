## 1. 거듭제곱: 2 × 2 × 2 = 2³

같은 수를 거듭 곱한 것을 **거듭제곱**이라고 해요.
점이 2배씩 늘어나는 모습으로 확인해 봅시다.

> 💡 위첨자는 키보드로 쓰기 어려우니 복사해서 쓰세요: ² ³ ⁴ ⁵

```python
%%manim -qm -v WARNING PowersOfTwo

class PowersOfTwo(Scene):
    def construct(self):
        title = Text("2를 거듭 곱하면?", font="NanumGothic").to_edge(UP)
        self.play(Write(title))

        formulas = ["2¹ = 2", "2² = 4", "2³ = 8", "2⁴ = 16"]

        for step in range(1, 5):                    # 1, 2, 3, 4
            count = 2 ** step                       # 2의 step제곱
            dots = VGroup(*[Dot(color=YELLOW) for _ in range(count)])
            dots.arrange_in_grid(rows=2, buff=0.4).shift(UP * 0.5)

            label = Text(formulas[step - 1], font="NanumGothic", font_size=36)
            label.shift(DOWN * 2.5)

            self.play(FadeIn(dots), Write(label))
            self.wait(0.8)
            self.play(FadeOut(dots), FadeOut(label))
        self.wait(1)
```

## 2. 소인수분해 가지 그림 (트리)

12를 소인수분해 하면? **12 = 2 × 6 = 2 × 2 × 3 = 2² × 3**

가지 그림이 자라나는 애니메이션을 만들어 봅시다.
소수(2, 3)는 주황색 동그라미로 표시할 거예요.

```python
%%manim -qm -v WARNING FactorTree

class FactorTree(Scene):
    def construct(self):
        # 숫자 위치 정하기: [x, y, 0]
        n12 = Text("12", font_size=48).move_to([0, 2.5, 0])
        n2a = Text("2", font_size=48, color=ORANGE).move_to([-1.5, 0.8, 0])
        n6  = Text("6", font_size=48).move_to([1.5, 0.8, 0])
        n2b = Text("2", font_size=48, color=ORANGE).move_to([0.5, -0.9, 0])
        n3  = Text("3", font_size=48, color=ORANGE).move_to([2.5, -0.9, 0])

        # 가지(선) 만들기: 부모에서 자식으로
        b1 = Line(n12.get_bottom(), n2a.get_top(), buff=0.15)
        b2 = Line(n12.get_bottom(), n6.get_top(), buff=0.15)
        b3 = Line(n6.get_bottom(), n2b.get_top(), buff=0.15)
        b4 = Line(n6.get_bottom(), n3.get_top(), buff=0.15)

        # 소수에는 동그라미!
        circles = VGroup(*[Circle(radius=0.45, color=ORANGE).move_to(m)
                           for m in [n2a, n2b, n3]])

        result = Text("12 = 2² × 3", font="NanumGothic", font_size=44, color=YELLOW)
        result.move_to([0, -2.6, 0])

        self.play(Write(n12))
        self.play(Create(b1), Create(b2), Write(n2a), Write(n6))
        self.play(Create(b3), Create(b4), Write(n2b), Write(n3))
        self.play(Create(circles))
        self.play(Write(result))
        self.wait(2)
```

## 3. 반복문으로 소수 찾기 — 에라토스테네스의 체

1부터 30까지 수 중에서 **소수**만 노란색으로 바뀌는 애니메이션이에요.
`is_prime` 함수가 소수인지 판별해 줍니다. (약수가 1과 자기 자신뿐인 수!)

```python
%%manim -qm -v WARNING PrimeFinder

def is_prime(n):
    if n < 2:
        return False
    for k in range(2, n):
        if n % k == 0:        # k로 나누어떨어지면 소수가 아님
            return False
    return True

class PrimeFinder(Scene):
    def construct(self):
        title = Text("1~30 중 소수는?", font="NanumGothic").to_edge(UP)
        self.play(Write(title))

        numbers = VGroup(*[Text(str(n), font_size=36) for n in range(1, 31)])
        numbers.arrange_in_grid(rows=5, cols=6, buff=0.6).shift(DOWN * 0.5)
        self.play(FadeIn(numbers))
        self.wait(0.5)

        for n in range(1, 31):
            mob = numbers[n - 1]                  # n이 적힌 글자
            if is_prime(n):
                self.play(mob.animate.set_color(YELLOW).scale(1.4),
                          run_time=0.25)
            else:
                self.play(mob.animate.set_color(GREY).scale(0.8),
                          run_time=0.1)
        self.wait(2)
```
