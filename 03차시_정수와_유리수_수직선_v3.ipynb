## 1. 수직선 그리기

`NumberLine`으로 수직선을 만들고, 반복문으로 눈금 숫자를 붙여 봅시다.

```python
%%manim -qm -v WARNING DrawNumberLine

class DrawNumberLine(Scene):
    def construct(self):
        line = NumberLine(x_range=[-6, 6, 1], length=12)

        # 반복문으로 -6 ~ 6 눈금 숫자 붙이기
        labels = VGroup()
        for i in range(-6, 7):
            t = Text(str(i), font_size=24)
            t.next_to(line.n2p(i), DOWN)     # n2p: 수 → 화면 위치
            labels.add(t)

        self.play(Create(line))
        self.play(FadeIn(labels))

        # 점 찍기: +3 은 파랑, -4 는 빨강, 1/2 은 초록
        p1 = Dot(line.n2p(3), color=BLUE).scale(1.3)
        p2 = Dot(line.n2p(-4), color=RED).scale(1.3)
        p3 = Dot(line.n2p(0.5), color=GREEN).scale(1.3)
        t3 = Text("1/2", font_size=24, color=GREEN).next_to(p3, UP)

        self.play(FadeIn(p1), FadeIn(p2), FadeIn(p3), Write(t3))
        self.wait(2)
```

## 2. 절댓값 = 0에서의 거리

|+3| = 3, |-3| = 3.
부호는 달라도 **0에서 떨어진 거리**는 같다는 걸 화살표로 보여줍시다.

```python
%%manim -qm -v WARNING AbsoluteValue

class AbsoluteValue(Scene):
    def construct(self):
        line = NumberLine(x_range=[-5, 5, 1], length=11)
        labels = VGroup(*[Text(str(i), font_size=24).next_to(line.n2p(i), DOWN)
                          for i in range(-5, 6)])
        self.play(Create(line), FadeIn(labels))

        zero = Dot(line.n2p(0), color=WHITE).scale(1.3)
        p_plus = Dot(line.n2p(3), color=BLUE).scale(1.3)
        p_minus = Dot(line.n2p(-3), color=RED).scale(1.3)
        self.play(FadeIn(zero), FadeIn(p_plus), FadeIn(p_minus))

        # 0에서 +3, -3까지의 거리를 화살표로
        arrow_r = DoubleArrow(line.n2p(0) + UP * 0.8, line.n2p(3) + UP * 0.8,
                              color=BLUE, buff=0)
        arrow_l = DoubleArrow(line.n2p(-3) + UP * 0.8, line.n2p(0) + UP * 0.8,
                              color=RED, buff=0)
        t_r = Text("|+3| = 3", font_size=30, color=BLUE).next_to(arrow_r, UP)
        t_l = Text("|-3| = 3", font_size=30, color=RED).next_to(arrow_l, UP)

        self.play(Create(arrow_r), Write(t_r))
        self.play(Create(arrow_l), Write(t_l))

        msg = Text("부호는 달라도 거리는 같다!", font="NanumGothic",
                   font_size=32, color=YELLOW).shift(DOWN * 2)
        self.play(Write(msg))
        self.wait(2)
```

## 3. 정수의 덧셈: (+3) + (-5) = ?

수직선 위에서
- **양수를 더하면** 오른쪽으로 이동 ➡️
- **음수를 더하면** 왼쪽으로 이동 ⬅️

점의 이동으로 (+3) + (-5) 를 계산해 봅시다.

```python
%%manim -qm -v WARNING IntegerAddition

class IntegerAddition(Scene):
    def construct(self):
        formula = Text("(+3) + (-5) = ?", font_size=40).to_edge(UP)
        line = NumberLine(x_range=[-6, 6, 1], length=12)
        labels = VGroup(*[Text(str(i), font_size=24).next_to(line.n2p(i), DOWN)
                          for i in range(-6, 7)])
        self.play(Write(formula), Create(line), FadeIn(labels))

        dot = Dot(line.n2p(0), color=YELLOW).scale(1.5)
        self.play(FadeIn(dot))

        # 1단계: 0에서 +3만큼 오른쪽으로
        arrow1 = Arrow(line.n2p(0) + UP * 0.7, line.n2p(3) + UP * 0.7,
                       color=BLUE, buff=0)
        t1 = Text("+3", font_size=30, color=BLUE).next_to(arrow1, UP)
        self.play(Create(arrow1), Write(t1))
        self.play(dot.animate.move_to(line.n2p(3)), run_time=1.5)

        # 2단계: +3에서 -5만큼 왼쪽으로
        arrow2 = Arrow(line.n2p(3) + UP * 1.5, line.n2p(-2) + UP * 1.5,
                       color=RED, buff=0)
        t2 = Text("-5", font_size=30, color=RED).next_to(arrow2, UP)
        self.play(Create(arrow2), Write(t2))
        self.play(dot.animate.move_to(line.n2p(-2)), run_time=1.5)

        answer = Text("(+3) + (-5) = -2", font_size=40, color=YELLOW)
        answer.to_edge(UP)
        self.play(Transform(formula, answer))
        self.wait(2)
```
