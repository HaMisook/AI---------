## 0. 스트링아트가 뭐예요?

못에 실을 규칙적으로 감으면 실은 분명 **곧은 직선**인데, 멀리서 보면 **곡선**이 보여요. 이렇게 직선의 모임이 만들어내는 곡선을 수학에서는 **포락선**이라고 불러요.

오늘은 두 개의 **샘플 코드**를 먼저 실행해 보고, 그다음 **실습 코드**에서 숫자를 바꿔 나만의 작품을 만들어요.


## 1. 샘플 ① — 직각 코너 포물선

가로 선분 위의 점과 세로 선분 위의 점을 **반대 순서**로 잇기만 했는데 포물선이 생겨요!

- 가로의 $i$번째 점 ↔ 세로의 $(n-i)$번째 점을 직선으로 연결
- 직선은 모두 곧지만, 그 직선들의 "경계"가 곡선이 됨

먼저 그대로 실행해서 모양을 확인하세요.


```python
%%manim -qm -v WARNING StringParabola

class StringParabola(Scene):
    def construct(self):
        n = 14            # 실(선분)의 개수
        L = 5             # 두 변의 길이
        origin = LEFT * 2.5 + DOWN * 2.5

        # 두 개의 기준 선분(가로축·세로축 역할)
        h_axis = Line(origin, origin + RIGHT * L, color=GREY)
        v_axis = Line(origin, origin + UP * L, color=GREY)
        self.play(Create(h_axis), Create(v_axis))

        strings = VGroup()
        for i in range(n + 1):
            a = origin + RIGHT * L * (i / n)        # 가로 위의 점
            b = origin + UP * L * ((n - i) / n)     # 세로 위의 점 (반대 순서!)
            strings.add(Line(a, b, stroke_width=2))
        strings.set_color_by_gradient(BLUE, PURPLE, PINK)

        self.play(LaggedStart(*[Create(s) for s in strings],
                              lag_ratio=0.15, run_time=4))
        self.wait(1)
```

## 2. 샘플 ② — 곱셈원 (배수의 무늬)

원 둘레에 점 $N$개를 번호 매겨 찍고, 각 점 $i$를 $(i\times k)$를 $N$으로 나눈 **나머지** 번호 점과 잇습니다.

- $k=2$ → 하트(심장형) 곡선
- $k=3$ → 콩팥형 곡선
- $k$를 바꿀 때마다 완전히 다른 무늬가 나와요!

곧은 실들이 모여 부드러운 곡선을 만드는 게 보일 거예요.


```python
%%manim -qm -v WARNING TimesTableCircle

class TimesTableCircle(Scene):
    def construct(self):
        N = 100           # 원 위의 점 개수
        k = 2             # 곱하는 수 (2면 하트, 3이면 콩팥 모양)
        R = 3.0           # 원의 반지름

        def point(i):
            # i번 점의 좌표 (맨 위에서 시작해 시계 방향)
            ang = PI / 2 - TAU * i / N
            return R * np.array([np.cos(ang), np.sin(ang), 0])

        circle = Circle(radius=R, color=GREY)
        label = Text(f"× {k}  (점 {N}개)", font="NanumGothic", font_size=30).to_edge(UP)
        self.play(Create(circle), Write(label))

        strings = VGroup()
        for i in range(N):
            j = (i * k) % N
            strings.add(Line(point(i), point(j), stroke_width=1.2))
        strings.set_color_by_gradient(TEAL, BLUE, PURPLE)

        self.play(Create(strings), run_time=5)
        self.wait(1)
```

## 3. 실습 — 나만의 스트링아트 만들기 🎨

아래 **실습 코드**에서 `N`, `k`, 색을 바꿔 보세요. 같은 코드인데 숫자 하나만 바꿔도 무늬가 확 달라져요.

추천 조합: `k=33, N=100` / `k=51, N=200` / `k=99, N=200`

> 💡 점이 많을수록(`N` 크게) 곡선이 더 매끄러워지지만 렌더링이 조금 느려져요.


```python
%%manim -qm -v WARNING MyStringArt

class MyStringArt(Scene):
    def construct(self):
        # ===== 여기 세 숫자를 바꿔 보세요 =====
        N = 120           # 점 개수
        k = 33            # 곱하는 수
        R = 3.0           # 반지름
        my_colors = [YELLOW, ORANGE, RED, PINK]   # 색도 바꿔 보세요
        # ====================================

        def point(i):
            ang = PI / 2 - TAU * i / N
            return R * np.array([np.cos(ang), np.sin(ang), 0])

        circle = Circle(radius=R, color=GREY)
        strings = VGroup()
        for i in range(N):
            j = (i * k) % N
            strings.add(Line(point(i), point(j), stroke_width=1.2))
        strings.set_color_by_gradient(*my_colors)

        self.play(Create(circle))
        self.play(Create(strings), run_time=5)
        self.wait(1)
```

## 🏆 도전 과제 — 움직이는 스트링아트 쇼

`k`가 자동으로 점점 커지면 무늬가 살아 움직이는 것처럼 변해요. 아래 셀을 실행해 보고, 시작·끝 `k` 값을 바꿔 나만의 "무늬 변신쇼"를 만들어 보세요.


```python
%%manim -qm -v WARNING StringArtShow

class StringArtShow(Scene):
    def construct(self):
        N = 120
        R = 3.0
        k = ValueTracker(2)        # k를 2 → 40 까지 움직일 거예요

        def point(i):
            ang = PI / 2 - TAU * i / N
            return R * np.array([np.cos(ang), np.sin(ang), 0])

        circle = Circle(radius=R, color=GREY)

        def make_strings():
            kk = k.get_value()
            g = VGroup()
            for i in range(N):
                j = (i * kk) % N
                g.add(Line(point(i), point(j), stroke_width=1.0))
            g.set_color_by_gradient(TEAL, BLUE, PURPLE, PINK)
            return g
        strings = always_redraw(make_strings)

        label = always_redraw(lambda: Text(
            f"× {k.get_value():.1f}", font="NanumGothic", font_size=30).to_edge(UP))

        self.add(circle, strings, label)
        self.play(k.animate.set_value(40), run_time=8, rate_func=linear)
        self.wait(1)
```

## 📝 오늘 배운 것 정리

- **곧은 직선**들이 규칙적으로 모이면 **곡선(포락선)** 이 보인다.
- 가로·세로 점을 반대 순서로 이으면 **포물선**이 생긴다.
- 원 위 점을 `(i×k) mod N` 규칙으로 이으면 **배수의 무늬**가 생긴다 (k=2 하트, k=3 콩팥).
- `for` 반복문 + `Line`, `set_color_by_gradient`, `ValueTracker`로 작품을 움직일 수 있다.
