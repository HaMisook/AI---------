## 1. 선분, 반직선, 직선

두 점 A, B가 있을 때:

- **선분 AB**: A에서 B까지 (양쪽 끝이 있음)
- **반직선 AB**: A에서 시작해 B 방향으로 끝없이
- **직선 AB**: 양쪽으로 끝없이 뻗음

```python
%%manim -qm -v WARNING LineTypes

class LineTypes(Scene):
    def construct(self):
        # 세 줄에 A, B 점 배치
        rows = [2, 0, -2]                 # 각 줄의 y좌표
        names = ["선분 AB", "반직선 AB", "직선 AB"]

        for y, name in zip(rows, names):
            A = Dot([-2, y, 0])
            B = Dot([1, y, 0])
            tA = Text("A", font_size=28).next_to(A, DOWN)
            tB = Text("B", font_size=28).next_to(B, DOWN)
            label = Text(name, font="NanumGothic", font_size=30).move_to([-5, y, 0])

            if name == "선분 AB":
                line = Line(A.get_center(), B.get_center(), color=YELLOW)
            elif name == "반직선 AB":
                line = Arrow([-2, y, 0], [3.5, y, 0], color=GREEN, buff=0)
            else:
                line = Arrow([-4, y, 0], [3.5, y, 0], color=BLUE, buff=0)
                line2 = Arrow([1, y, 0], [-4.5, y, 0], color=BLUE, buff=0)

            self.play(Write(label), FadeIn(A), FadeIn(B), Write(tA), Write(tB))
            self.play(Create(line))
            if name == "직선 AB":
                self.play(Create(line2))
        self.wait(2)
```

## 2. 각 그리기

각은 한 점(꼭짓점)에서 시작하는 두 반직선이 만들어요.
`Angle(선1, 선2)`로 각 표시를 그릴 수 있습니다.

> `PI`는 180°예요. 60°는 `60 * DEGREES`로 쓰면 편해요!

```python
%%manim -qm -v WARNING DrawAngle

import numpy as np

class DrawAngle(Scene):
    def construct(self):
        O = ORIGIN                                   # 꼭짓점
        ray1 = Line(O, RIGHT * 3)                    # 0° 방향 반직선
        ray2 = Line(O, 3 * np.array([np.cos(60 * DEGREES),
                                     np.sin(60 * DEGREES), 0]))  # 60° 방향

        angle_mark = Angle(ray1, ray2, radius=0.7, color=YELLOW)
        label = Text("60°", font_size=32, color=YELLOW).move_to([1.3, 0.65, 0])

        vertex = Dot(O)
        tO = Text("O", font_size=28).next_to(vertex, DOWN + LEFT)

        self.play(FadeIn(vertex), Write(tO))
        self.play(Create(ray1))
        self.play(Create(ray2))
        self.play(Create(angle_mark), Write(label))
        self.wait(1)

        # 직각(90°)도 그려 보기 — RightAngle은 ㄱ자 모양 표시
        ray_up = Line(O, UP * 3)
        self.play(FadeOut(angle_mark), FadeOut(label),
                  Transform(ray2, ray_up))
        right_mark = RightAngle(ray1, ray_up, length=0.5, color=GREEN)
        label90 = Text("90°", font_size=32, color=GREEN).move_to([1.0, 1.0, 0])
        self.play(Create(right_mark), Write(label90))
        self.wait(2)
```

## 3. 맞꼭지각: 두 직선이 만나면?

두 직선이 한 점에서 만날 때 **마주 보는 두 각(맞꼭지각)** 의 크기는 항상 같아요.
색깔로 확인해 봅시다.

```python
%%manim -qm -v WARNING VerticalAngles

import numpy as np

class VerticalAngles(Scene):
    def construct(self):
        # 두 직선: 30° 방향과 110° 방향
        d1 = np.array([np.cos(30 * DEGREES), np.sin(30 * DEGREES), 0])
        d2 = np.array([np.cos(110 * DEGREES), np.sin(110 * DEGREES), 0])

        line1 = Line(-4 * d1, 4 * d1, color=WHITE)
        line2 = Line(-4 * d2, 4 * d2, color=WHITE)
        self.play(Create(line1), Create(line2))

        # 맞꼭지각 한 쌍 (노랑): 오른쪽 위 & 왼쪽 아래
        a1 = Angle(Line(ORIGIN, 4 * d1), Line(ORIGIN, 4 * d2),
                   radius=0.8, color=YELLOW)
        a2 = Angle(Line(ORIGIN, -4 * d1), Line(ORIGIN, -4 * d2),
                   radius=0.8, color=YELLOW)
        # 나머지 한 쌍 (파랑)
        b1 = Angle(Line(ORIGIN, 4 * d2), Line(ORIGIN, -4 * d1),
                   radius=0.6, color=BLUE)
        b2 = Angle(Line(ORIGIN, -4 * d2), Line(ORIGIN, 4 * d1),
                   radius=0.6, color=BLUE)

        self.play(Create(a1), Create(a2))
        self.play(Create(b1), Create(b2))

        msg = Text("마주 보는 각(맞꼭지각)의 크기는 같다!",
                   font="NanumGothic", font_size=32, color=YELLOW)
        msg.to_edge(DOWN)
        self.play(Write(msg))
        self.wait(2)
```

## 🏆 도전 과제

1. `DrawAngle` 코드를 고쳐서 **30°, 120°** 각을 그려 보세요. (`60 * DEGREES`를 바꾸면 돼요. 라벨 위치도 옮겨야겠죠?)
2. (보너스) 시계가 3시일 때 시침과 분침이 이루는 각을 그려 보세요. (몇 도일까요?)

```python
%%manim -qm -v WARNING MyAngle

import numpy as np

class MyAngle(Scene):
    def construct(self):
        O = ORIGIN
        ray1 = Line(O, RIGHT * 3)
        # 각도를 바꿔 보세요!
        deg = 60
        ray2 = Line(O, 3 * np.array([np.cos(deg * DEGREES),
                                     np.sin(deg * DEGREES), 0]))

        angle_mark = Angle(ray1, ray2, radius=0.7, color=YELLOW)
        label = Text(str(deg) + "°", font_size=32, color=YELLOW)
        label.next_to(angle_mark, RIGHT + UP, buff=0.1)

        self.play(Create(ray1), Create(ray2))
        self.play(Create(angle_mark), Write(label))
        self.wait(2)
```
