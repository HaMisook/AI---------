## 1. 작도란?

**눈금 없는 자**와 **컴퍼스**만으로 도형을 그리는 것!
코드에서는

- 컴퍼스로 원 그리기 → `Circle(radius=r).move_to(중심)`
- 자로 선 긋기 → `Line(점1, 점2)`

### 선분 AB의 수직이등분선 작도 순서
1. A를 중심으로 반지름이 충분히 큰 원을 그린다.
2. B를 중심으로 **같은 반지름**의 원을 그린다.
3. 두 원이 만나는 두 점을 직선으로 잇는다. → 이 직선이 수직이등분선!

```python
%%manim -qm -v WARNING PerpendicularBisector

import numpy as np

class PerpendicularBisector(Scene):
    def construct(self):
        A = np.array([-2, 0, 0])
        B = np.array([2, 0, 0])
        r = 3                                   # 컴퍼스 반지름

        dotA, dotB = Dot(A), Dot(B)
        tA = Text("A", font_size=28).next_to(dotA, DOWN + LEFT)
        tB = Text("B", font_size=28).next_to(dotB, DOWN + RIGHT)
        seg = Line(A, B)
        self.play(FadeIn(dotA), FadeIn(dotB), Write(tA), Write(tB), Create(seg))

        # 1, 2단계: 컴퍼스로 원 두 개 (점선으로 표시)
        c1 = DashedVMobject(Circle(radius=r, color=BLUE).move_to(A))
        c2 = DashedVMobject(Circle(radius=r, color=GREEN).move_to(B))
        self.play(Create(c1))
        self.play(Create(c2))

        # 두 원의 교점 계산: x=0, y=±√(r²-2²)
        h = np.sqrt(r ** 2 - 2 ** 2)
        P = np.array([0, h, 0])
        Q = np.array([0, -h, 0])
        dotP, dotQ = Dot(P, color=YELLOW), Dot(Q, color=YELLOW)
        self.play(FadeIn(dotP), FadeIn(dotQ))

        # 3단계: 두 교점을 잇는 직선
        bisector = Line(P * 1.4, Q * 1.4, color=YELLOW)
        self.play(Create(bisector))

        # 확인: 중점과 직각 표시
        M = Dot(ORIGIN, color=RED)
        mark = RightAngle(Line(ORIGIN, B), Line(ORIGIN, P),
                          length=0.3, color=RED)
        msg = Text("수직 + 이등분!", font="NanumGothic",
                   font_size=32, color=YELLOW).to_edge(DOWN)
        self.play(FadeIn(M), Create(mark), Write(msg))
        self.wait(2)
```

## 2. 각의 이등분선 작도

### 작도 순서
1. 꼭짓점 O를 중심으로 원을 그려 두 변과 만나는 점 P, Q를 찾는다.
2. P, Q를 중심으로 같은 반지름의 원을 그려 교점 X를 찾는다.
3. O와 X를 잇는다. → 각이 **정확히 반으로**!

```python
%%manim -qm -v WARNING AngleBisector

import numpy as np

class AngleBisector(Scene):
    def construct(self):
        O = np.array([-3, -2, 0])
        deg = 60                                       # 전체 각
        d1 = np.array([1, 0, 0])                       # 0° 방향
        d2 = np.array([np.cos(deg * DEGREES), np.sin(deg * DEGREES), 0])

        ray1 = Line(O, O + 6 * d1)
        ray2 = Line(O, O + 6 * d2)
        self.play(Create(ray1), Create(ray2))

        # 1단계: O 중심 원 → 교점 P, Q
        r1 = 2.0
        arc1 = Arc(radius=r1, start_angle=-15 * DEGREES,
                   angle=(deg + 30) * DEGREES, arc_center=O, color=BLUE)
        P = O + r1 * d1
        Q = O + r1 * d2
        self.play(Create(arc1))
        self.play(FadeIn(Dot(P, color=BLUE)), FadeIn(Dot(Q, color=BLUE)))

        # 2단계: P, Q 중심의 원 → 교점 X (이등분선 방향, 거리 계산)
        r2 = 1.6
        half = deg / 2
        mid_dir = np.array([np.cos(half * DEGREES), np.sin(half * DEGREES), 0])
        chord_half = r1 * np.sin(half * DEGREES)       # PQ의 절반
        dist = r1 * np.cos(half * DEGREES) + np.sqrt(r2 ** 2 - chord_half ** 2)
        X = O + dist * mid_dir

        arcP = Arc(radius=r2, start_angle=(half + 90) * DEGREES,
                   angle=-120 * DEGREES, arc_center=P, color=GREEN)
        arcQ = Arc(radius=r2, start_angle=(half - 90) * DEGREES,
                   angle=120 * DEGREES, arc_center=Q, color=GREEN)
        self.play(Create(arcP), Create(arcQ))
        self.play(FadeIn(Dot(X, color=YELLOW)))

        # 3단계: O와 X 잇기
        bisector = Line(O, O + 6 * mid_dir, color=YELLOW)
        self.play(Create(bisector))

        # 반으로 나뉜 두 각 표시
        a1 = Angle(ray1, bisector, radius=0.8, color=ORANGE)
        a2 = Angle(bisector, ray2, radius=0.9, color=PINK)
        t1 = Text("30°", font_size=26, color=ORANGE)
        t1.move_to(O + 1.4 * np.array([np.cos(15 * DEGREES), np.sin(15 * DEGREES), 0]))
        t2 = Text("30°", font_size=26, color=PINK)
        t2.move_to(O + 1.5 * np.array([np.cos(45 * DEGREES), np.sin(45 * DEGREES), 0]))
        self.play(Create(a1), Create(a2), Write(t1), Write(t2))
        self.wait(2)
```

## 3. 정삼각형 작도 — 두 원의 마법

선분 AB가 있을 때, A 중심·반지름 AB인 원과 B 중심·반지름 AB인 원의
**교점 C**를 찾으면 △ABC는 정삼각형! 왜일까요?

→ AC도 반지름, BC도 반지름이므로 **AB = AC = BC** 이기 때문이에요.

```python
%%manim -qm -v WARNING EquilateralTriangle

import numpy as np

class EquilateralTriangle(Scene):
    def construct(self):
        A = np.array([-1.5, -1, 0])
        B = np.array([1.5, -1, 0])
        r = 3                                    # AB의 길이 = 반지름

        seg = Line(A, B)
        self.play(Create(seg), FadeIn(Dot(A)), FadeIn(Dot(B)))

        c1 = DashedVMobject(Circle(radius=r, color=BLUE).move_to(A))
        c2 = DashedVMobject(Circle(radius=r, color=GREEN).move_to(B))
        self.play(Create(c1))
        self.play(Create(c2))

        # 교점 C: AB 중점 위쪽 √3/2 × r
        C = np.array([0, -1 + r * np.sqrt(3) / 2, 0])
        dotC = Dot(C, color=YELLOW)
        self.play(FadeIn(dotC))

        triangle = Polygon(A, B, C, color=YELLOW)
        triangle.set_fill(YELLOW, opacity=0.3)
        self.play(Create(triangle), FadeOut(c1), FadeOut(c2))

        msg = Text("세 변이 모두 반지름 → 정삼각형!",
                   font="NanumGothic", font_size=32, color=YELLOW).to_edge(DOWN)
        self.play(Write(msg))
        self.wait(2)
```
