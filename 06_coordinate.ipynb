## 1. 좌표평면과 순서쌍

`NumberPlane`으로 모눈종이(좌표평면)를 만들고,
`c2p(x, y)` (coordinate to point)로 순서쌍의 화면 위치를 구해요.

```python
%%manim -qm -v WARNING CoordinateBasics

class CoordinateBasics(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6, 1], y_range=[-4, 4, 1],
            background_line_style={"stroke_opacity": 0.4}
        )
        self.play(Create(plane))

        # 점 A(3, 2) 찍기 — 가는 길을 화살표로 안내
        origin = Dot(plane.c2p(0, 0), color=WHITE)
        path_x = Arrow(plane.c2p(0, 0), plane.c2p(3, 0), color=BLUE, buff=0)
        path_y = Arrow(plane.c2p(3, 0), plane.c2p(3, 2), color=GREEN, buff=0)
        A = Dot(plane.c2p(3, 2), color=YELLOW).scale(1.3)
        tA = Text("A(3, 2)", font_size=30, color=YELLOW).next_to(A, UP + RIGHT)

        self.play(FadeIn(origin))
        self.play(Create(path_x))      # x축 방향으로 3
        self.play(Create(path_y))      # y축 방향으로 2
        self.play(FadeIn(A), Write(tA))
        self.wait(0.5)

        # 순서가 바뀌면 다른 점! B(2, 3)
        B = Dot(plane.c2p(2, 3), color=PINK).scale(1.3)
        tB = Text("B(2, 3)", font_size=30, color=PINK).next_to(B, UP + LEFT)
        self.play(FadeIn(B), Write(tB))

        msg = Text("(3,2)와 (2,3)은 다른 점! 순서가 중요해요",
                   font="NanumGothic", font_size=28, color=YELLOW).to_edge(DOWN)
        self.play(Write(msg))
        self.wait(2)
```

## 2. 네 개의 사분면

좌표평면은 x축, y축에 의해 네 부분으로 나뉘어요.

| 사분면 | x부호 | y부호 |
|---|---|---|
| 제1사분면 | + | + |
| 제2사분면 | − | + |
| 제3사분면 | − | − |
| 제4사분면 | + | − |

```python
%%manim -qm -v WARNING Quadrants

class Quadrants(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6, 1], y_range=[-4, 4, 1],
            background_line_style={"stroke_opacity": 0.4}
        )
        self.play(Create(plane))

        # (이름, 점의 좌표, 색) 목록 → 반복문으로 한 번에!
        data = [
            ("제1사분면 (+,+)", (3, 2), YELLOW),
            ("제2사분면 (-,+)", (-3, 2), GREEN),
            ("제3사분면 (-,-)", (-3, -2), BLUE),
            ("제4사분면 (+,-)", (3, -2), PINK),
        ]

        for name, (x, y), color in data:
            dot = Dot(plane.c2p(x, y), color=color).scale(1.3)
            label = Text(name, font="NanumGothic", font_size=26, color=color)
            label.next_to(dot, UP)
            coord = Text(f"({x}, {y})", font_size=24, color=color)
            coord.next_to(dot, DOWN)
            self.play(FadeIn(dot), Write(label), Write(coord), run_time=0.8)
        self.wait(2)
```

## 3. 좌표로 그림 그리기 ⭐

좌표 목록을 순서대로 선분으로 이으면 그림이 돼요!
아래는 별 모양 좌표예요. 반복문이 점을 하나씩 이어 줍니다.

```python
%%manim -qm -v WARNING StarPicture

class StarPicture(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6, 1], y_range=[-4, 4, 1],
            background_line_style={"stroke_opacity": 0.4}
        )
        self.play(Create(plane))

        # 별을 만드는 순서쌍 목록 (마지막에 처음 점을 한 번 더!)
        coords = [(0, 3), (1, 1), (3, 1), (1.5, -0.5),
                  (2, -2.5), (0, -1), (-2, -2.5), (-1.5, -0.5),
                  (-3, 1), (-1, 1), (0, 3)]

        prev = None
        for x, y in coords:
            p = plane.c2p(x, y)
            dot = Dot(p, color=YELLOW, radius=0.06)
            self.play(FadeIn(dot), run_time=0.15)
            if prev is not None:
                self.play(Create(Line(prev, p, color=YELLOW)), run_time=0.3)
            prev = p
        self.wait(2)
```

## 🏆 도전 과제

1. `StarPicture`의 `coords` 목록을 바꿔서 **자기만의 그림**(집, 번개, 알파벳 등)을 그려 보세요.
2. 종이 모눈에 먼저 스케치하고 좌표를 읽어 오면 쉬워요!
3. (보너스) 색을 중간에 바꾸거나, `dot`에 라벨을 붙여 보세요.

```python
%%manim -qm -v WARNING MyPicture

class MyPicture(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6, 1], y_range=[-4, 4, 1],
            background_line_style={"stroke_opacity": 0.4}
        )
        self.play(Create(plane))

        # 좌표를 바꿔서 나만의 그림을 그려 보세요!
        coords = [(0, 0), (2, 0), (2, 2), (0, 2), (0, 0)]

        prev = None
        for x, y in coords:
            p = plane.c2p(x, y)
            self.play(FadeIn(Dot(p, color=GREEN, radius=0.06)), run_time=0.15)
            if prev is not None:
                self.play(Create(Line(prev, p, color=GREEN)), run_time=0.3)
            prev = p
        self.wait(2)
```

## 📝 오늘 배운 것 정리

- `NumberPlane` + `c2p(x, y)` 로 좌표평면 위 위치를 구한다.
- 순서쌍은 순서가 중요하다: (3,2) ≠ (2,3)
- 사분면별 부호: (+,+), (−,+), (−,−), (+,−)
- 좌표 목록 + 반복문 = 그림!
